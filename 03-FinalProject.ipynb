## Abstract

In order to understand how the relationship between NBA player height and three-point shooting behavior changed over time, we conducted an exploratory and statistical analysis covering three distinct eras of professional basketball. Our research was driven by the dramatic strategic transformation the NBA has undergone over the past 30 years; from a big man dominated era in the 1990s to the current analytics-driven game where three-point shooting is required of players at every position. In light of this change, we aimed to provide observational findings on whether height, which has long been thought to be a crucial factor in determining a player’s role and shot selection, still predicts how often a player attempts three-pointers and whether that relationship has changed over time. 

Using two endpoints, LeagueDashPlayerStats for season-level box score data and PlayerIndex for player height and positions metadata, we collected player statistics and biographical information from the official NBA Stats API using the nba_api Python module. After filtering for significant playing time, we acquired almost 6,000 player-season observations from 8 representative seasons across three eras (1990-2000, 2001-2013, and 2014-present). We selected FG3A_RATE, which is defined as three-point attempts as a percentage of all shot attempts, as our main variable of interest since it can represent shot selection independent of playing time and era-to-era pace differences. We grouped players into three positional categories (Guards, Wings, and Bigs) and applied linear regression within each era to determine the direction, strength, and statistical significance. 

Our analysis’s findings showed that player height and three-point attempt rate have a negative correlation that is highly statistically significant in each of the three eras. However, the league as a whole has shifted dramatically upward in three-point shooting over time. Most shockingly, from 1996 to 2024, FG3A_RATE increased by 1906% for Bigs (centers and power forwards) who historically attempted very few three-pointers, compared to around 200% for Guards and Wings. The absolute difference between shorter and taller players remained the same despite this, and the regression slope was consistently negative during all three eras since Guards and Wings also significantly raised their rates. 

Our study concludes by clarifying the evolving yet persistent relationship between professional basketball players’ height and three-point shooting. Although height is still a statistically significant predictor of three-point attempt rate, role, position, and era each play an important part on a player’s shot selection. Most notably, the modern NBA seems to be splitting, creating a new “stretch big” archetype who shoots from distance alongside traditional interior players who do not. This study contributes to the broader people of sports analytics and player role evaluation, showing insights for coaches, analysts, and fans interested in how the game continues to evolve.

## Research Question

How has the correlation between height and 3-point attempts changed across select NBA eras? Furthermore, we can then pursue another question: Is the increase in 3-point shooting volume only apply to taller positions (power forwards and centers) or does this affect all positions equally?

Defining eras:

- ***Era 1 (1990-2000 seasons):*** Big-man dominated era where most points were scored within close distances to the basket as opposed to three-pointers (much further away from the basket).

- ***Era 2 (2001-2013 seasons):*** League begins to shy away from traditional post-up and big-men due to the removal of the "illegal defense" which demanded players must be glued to the player they are defending at all times. With the abolishment of that rule, teams were allowed to play zone defense, which allowed players to occupy space on the court and prevent the other team from taking it instead of solely sticking to the player. In zone defenses, teams could theoretically just sufforcate the paint with their players and prevent anyone from making their way into the basket, taking away points from closer to the basket. To counter this, teams began to space out their players in the court, in attempts to work with the little space they have in the perimters to score from further away the basket. 

- ***Era 3 (2014-Present seasons):*** Teams analyzed shot attempts throughout their seasons and come to the realization that mid-range shots just are not as efficient as shooting 3-pointers. The chances of missing a mid-range and a three-pointer are the same, but three-pointers reward more points than a mid-range, meaning teams would much rather shoot 3's going forward. Teams now push all their players to attempt shooting 3-pointers, rather than mid-range jumpers, further increasing the volume of 3-pointers in the league.


## Background and Prior Work

The National Basketball Association (NBA) has experienced a major strategic shift over the past several decades, most visibly through the rapid rise of the three-point shot. Ever since the introduction of the three-point line in 1979, teams have increasingly emphasized long-range shooting as a high-value offensive strategy (initially at an extreme pace, but continuing at a solid pace well into the modern "pace-and-space" era).<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)
League-wide data show that three-point attempts per game have increased dramatically since the early 2000s, reshaping offensive decision-making and redefining traditional player roles, both of which are crucially linked to player height.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

Historically, there was a clear negative relationship between player height and shooting range. Guards, the shortest players on the court, were primarily responsible for perimeter scoring and playmaking, while centers and power forwards operated near the basket. Research by Podmenik et al. in the Journal of Human Kinetics supports this, showing that point guards and shooting guards “shoot more frequently from longer distances compared to players of other positions” due to their roles and biomechanical adjustments.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Similarly, analysis from RunRepeat shows that players shorter than the league average consistently attempted more long-range shots than taller players, reinforcing the idea that height limited three-point opportunities for big men because of their responsibilities in the paint.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Together, these findings suggest that for much of NBA history, taller players were discouraged from shooting from distance, creating a consistent inverse relationship between height and three-point attempts.

On the other hand, research from the Berkeley Sports Analytics Group explores positional differences in three-point shooting, showing that guards historically attempted far more three-pointers than forwards or centers, but that this disparity has narrowed substantially in recent seasons.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)
This finding potentially suggests that traditional positional expectations are becoming less rigid, with forwards and centers who are generally considered to be taller players increasingly expected to contribute as perimeter shooters. While this work addresses position and era, it does not directly isolate player height as a continuous variable, leaving open the question of how stature itself correlates with three-point attempt frequency. Adding on, research on the modern era of “positionless basketball” shows that taller players are increasingly taking perimeter shots, a role that was once exclusive to guards. Zając et al., writing in the International Journal of Environmental Research and Public Health, document the rise of the “stretch five,” noting that modern post players now have the shooting skills and training needed to score effectively from beyond the arc.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) This evidence suggests that while the historical negative correlation between height and three-point attempts was strong, recent changes in strategy and player development may have weakened this relationship in the last decade.

Prior analytical work has also examined how this rise in three-point shooting varies across eras and positions. The UCLA DataRes analysis of NBA shooting performance demonstrates that the increase in three-point attempts is not uniform over time, but rather with especially sharp growth occurring after rule changes and strategic innovations that favored floor spacing and shot efficiency.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)
Their breakdown highlights how changes in league philosophy rather than individual player talent alone have driven shifts in shot selection. Additionally, the DataRes group does investigate a similar problem as we attempt, namely how height in modern players affect their choice of shot. However, their investigation omits a combination of temporal, positional, and physical elements and rather only focuses on the relationship between distance of shots and heights of players in the 2023-2024 season.

At a more foundational level, Erčulj and Štrumbelj (2015) provide a detailed empirical analysis of basketball shot selection and shot success across multiple levels of competition, including the NBA, Euroleague, and youth leagues.<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6)
Using manual video annotation of over 5,000 non-free-throw shots, the authors categorize shots by type (e.g., above-head shots, layups, dunks), player type (guard, forward, center), and situational variables such as shot distance and transition play. Their results show that guards attempt more long-range shots than centers, while taller player types are more likely to attempt shots near the basket such as dunks and tip-ins (note that we are specifically referring to specific positions as being "taller" or "shorter" "player types" as opposed to incorrectly directly interpreting them as tall or short players).

While these studies clearly demonstrate that all three elements of player type, physical characteristics, and NBA eras influence shot selection, they do so in an incomplete manner. There is a distinct lack of examination of player height as a continuous variable as well as a lack of specific focus on three-point attempt frequency across historical eras. This leaves an open gap in the literature regarding how height interacts with evolving positional roles and long-range shooting trends over time, which is what drives our project’s focus on height, position, and era-specific three-point behavior.

```python
# Run this everytime you load this notebook so the necessary functions can be performed.

import os
import time
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Setup + reproducible data pull / wrangling scaffold

try:
    from nba_api.stats.endpoints import leaguedashplayerstats, playerindex
except ImportError as exc:
    raise ImportError("Missing dependency `nba_api` in the active notebook kernel.") from exc

warnings.filterwarnings("ignore", category=FutureWarning)

SEASONS = [
    "1996-97", "1999-00", "2004-05", "2008-09", "2012-13", "2016-17", "2020-21", "2023-24"
]
REQUEST_TIMEOUT = 120
MAX_RETRIES = 4
BASE_BACKOFF_SECONDS = 2


def season_to_era(season: str) -> str:
    start_year = int(season.split("-")[0])
    if 1990 <= start_year <= 2000:
        return "1990-2000"
    if 2001 <= start_year <= 2013:
        return "2001-2013"
    return "2014-present"


def parse_height_to_inches(height_value):
    if pd.isna(height_value):
        return np.nan
    if isinstance(height_value, (int, float)):
        return float(height_value)

    s = str(height_value).strip().replace('"', "")
    if "-" in s:
        feet, inches = s.split("-", 1)
        if feet.isdigit() and inches.isdigit():
            return int(feet) * 12 + int(inches)

    try:
        return float(s)
    except ValueError:
        return np.nan


def standardize_position(position_value) -> str:
    if pd.isna(position_value):
        return "Unknown"
    p = str(position_value).upper()
    if "C" in p:
        return "Big"
    if "F" in p:
        return "Wing"
    if "G" in p:
        return "Guard"
    return "Unknown"


def fetch_with_retry(fetch_fn, label: str, retries: int = MAX_RETRIES):
    last_error = None
    for attempt in range(1, retries + 1):
        try:
            return fetch_fn()
        except Exception as exc:
            last_error = exc
            if attempt < retries:
                wait_s = BASE_BACKOFF_SECONDS ** attempt
                print(f"[{label}] attempt {attempt}/{retries} failed: {exc}")
                print(f"Retrying in {wait_s}s...")
                time.sleep(wait_s)
    raise RuntimeError(f"Failed to fetch {label} after {retries} attempts") from last_error


# Pull player metadata and include historical players
player_meta = fetch_with_retry(
    lambda: playerindex.PlayerIndex(historical_nullable=1, timeout=REQUEST_TIMEOUT).get_data_frames()[0].copy(),
    label="PlayerIndex",
)
id_col = "PERSON_ID" if "PERSON_ID" in player_meta.columns else "PLAYER_ID"
height_col = "HEIGHT" if "HEIGHT" in player_meta.columns else "PLAYER_HEIGHT"
position_col = "POSITION" if "POSITION" in player_meta.columns else "PLAYER_POSITION"

meta_cols = [id_col, height_col, position_col]
for optional_col in ["FROM_YEAR", "TO_YEAR"]:
    if optional_col in player_meta.columns:
        meta_cols.append(optional_col)

player_meta = player_meta[meta_cols].rename(
    columns={
        id_col: "PLAYER_ID",
        height_col: "HEIGHT",
        position_col: "POSITION",
    }
)

# Normalize join key types before merge
player_meta["PLAYER_ID"] = pd.to_numeric(player_meta["PLAYER_ID"], errors="coerce").astype("Int64")
player_meta = player_meta.dropna(subset=["PLAYER_ID"]).drop_duplicates(subset=["PLAYER_ID"], keep="last")

# Pull season-level stats for selected seasons
season_frames = []
for season in SEASONS:
    print(f"Pulling season {season}...")
    sdf = fetch_with_retry(
        lambda s=season: leaguedashplayerstats.LeagueDashPlayerStats(
            season=s,
            season_type_all_star="Regular Season",
            per_mode_detailed="Totals",
            timeout=REQUEST_TIMEOUT,
        ).get_data_frames()[0],
        label=f"LeagueDashPlayerStats {season}",
    )

    required_cols = ["PLAYER_ID", "PLAYER_NAME", "GP", "MIN", "FG3A", "FGA"]
    keep_cols = [c for c in required_cols if c in sdf.columns]
    sdf = sdf[keep_cols].copy()
    sdf["PLAYER_ID"] = pd.to_numeric(sdf["PLAYER_ID"], errors="coerce").astype("Int64")
    sdf["SEASON"] = season
    season_frames.append(sdf)
    time.sleep(1.0)

season_stats = pd.concat(season_frames, ignore_index=True)
season_stats = season_stats.dropna(subset=["PLAYER_ID"])

# Merge + feature engineering
df = season_stats.merge(player_meta, on="PLAYER_ID", how="left")

merge_success = df["HEIGHT"].notna().mean()
print(f"Metadata merge success (non-null HEIGHT): {merge_success:.1%}")

# Fallback if merge quality is unexpectedly low
if merge_success < 0.20:
    raise RuntimeError(
        "Metadata merge success is too low (<20%). Check NBA API response schema or connectivity."
    )

df["HEIGHT_IN"] = df["HEIGHT"].apply(parse_height_to_inches)
df["POSITION_GROUP"] = df["POSITION"].apply(standardize_position)
df["ERA"] = df["SEASON"].apply(season_to_era)

df["MPG"] = df["MIN"] / df["GP"].replace(0, np.nan)
df["FG3A_PER_GAME"] = df["FG3A"] / df["GP"].replace(0, np.nan)
df["FG3A_PER36"] = df["FG3A"] / (df["MIN"] / 36).replace(0, np.nan)
df["FG3A_RATE"] = df["FG3A"] / df["FGA"].replace(0, np.nan)

# Data quality filters from project plan
df = df[(df["GP"] >= 41) & (df["MPG"] >= 10)].copy()

# Save outputs for reproducibility
os.makedirs("data/01-interim", exist_ok=True)
os.makedirs("data/02-processed", exist_ok=True)

interim_path = "data/01-interim/nba_player_season_combined.csv"
processed_path = "data/02-processed/nba_height_3pa_player_season.csv"

df.to_csv(interim_path, index=False)
df.to_csv(processed_path, index=False)

print(f"Saved interim dataset: {interim_path}")
print(f"Saved processed dataset: {processed_path}")
print(f"Rows: {len(df):,} | Columns: {df.shape[1]}")

df.head()
```

First we observed the slopes of the 3 eras' regression lines and saw that the slopes were relatively similar which meant that the slopes of the regression lines that outline the height to rate of FG3PA throughout the eras has remained at similar levels. From earlier analysis, we discovered that the rate of FG3PA has changed drastically in the modern era, where we see more bigs shooting 3-pointers at a higher volume than any other era. To quantify this change, we observed the median rates of FG3PA of bigs for all 8 seasons, and identified a significant spike in 3-point volume starting in the 2020 saeson, jumping from a mere 0.02 to a 0.08, and spiking even further to a 0.17 in just 3 years. If the 3-point volume spiked so drastically, how is it that the slope of the modern era's regression line remain consistent with the previous eras. This is achieved through the balancing rate of FG3PA from guards and wings, which was only possible because their rates did not see a dramatic increase, unlike bigs. 

```python
for group in ['Guard', 'Wing', 'Big']:
    first = wrangled_df[(wrangled_df['POSITION_GROUP'] == group) & (wrangled_df['SEASON'] == '1996-97')]['FG3A_RATE'].median()
    last = wrangled_df[(wrangled_df['POSITION_GROUP'] == group) & (wrangled_df['SEASON'] == '2023-24')]['FG3A_RATE'].median()
    pct_change = ((last - first) / first) * 100
    print(f"{group}: {first:.2f} to {last:.2f} which is a {pct_change:.2f}% change.")
```

In the previous code cell, by calculating the median rates of FG3PA for guards, wings, and bigs, and then comparing it to the rates seen in the modern-era, we are able to identify that bigs in fact saw the most drastic change at a 1906% change, where as guards and wings, saw no more than a 200% increase. This meant that the magnitude of change in rates of FG3PA for bigs compared to wings was 9.5 times, and 54 times for guards.

## Discusison and Conclusion

We started by analyzing players’ general season statistics, playing time, and physical traits, which helped establish the overall structure of the dataset before focusing more directly on three-point behavior. The EDA suggests that height and three-point shooting behave very differently in our dataset. Height is relatively stable and follows a clear overall structure, with most players clustered near the league average and with noticeable separation between guards, wings, and bigs. By contrast, three-point attempt behavior is much more uneven and changes substantially across eras, with many players in earlier seasons attempting very few threes and much higher rates appearing in modern seasons. This helps explain our main result: height remains an important predictor of shot selection, but the strategic role of three-point shooting has changed so much over time that the same physical differences now exist within a very different offensive environment. 

Overall, our results mostly support the original hypothesis that taller NBA players attempt fewer three-point shots than shorter players. Across all three eras, there exists a negative relationship between height and three-point attempts (3PAs), meaning that as height increases, 3PA rates show a downwards trend. This matches the general idea that shorter players, especially guards, have traditionally played more on the perimeter, while taller players have usually taken on more inside roles. At the same time, our results also show that three-point attempt rates have gone up across all position groups in the modern NBA, suggesting that the increase in 3PA rates is a phenomenon affecting the whole NBA league and not just smaller players: Guards still have the highest median 3PA rate, but wings have increased a lot, and bigs show the most noticeable jump compared to earlier eras. As such, it seems accurate to conclude that the modern NBA has made player roles more flexible, while maintaining height as an important factor in shot selection.

In addition, our observation that the median rates of 3PA of bigs spiked starting in the 2020 season (0.02 to a 0.08, and spiking further to a 0.17 within 3 years) is indication that there has been a dramatic shift in offensive decision-making and the philosophy behind player roles in the modern NBA climate.

Our findings appear to be in agreement with what background research indicates, also having observed the idea that modern games (later eras and seasons) show increased 3PA rates as well as being in unison with the idea that guards have historically attempted more long-range shots than taller players, while also extending that idea into the modern era. While we also confirm the Berkeley Sports Analytics Group’s research pointing to a narrowing disparity in 3PA rates between positions, our examination additionally indicates that when grouped by height, 3PA rate disparity is decreasing too (2014-present era displaying a more level slope than the two previous eras). 

However, there are also several limitations to our project. Since we utilized observational data, we cannot make a causal claim that height directly determines three-point attempt behavior. Other factors such as team system, coaching strategy, player skill set, and role on offense likely influence shot selection too. Also, our position groups are broad, which may overlook important differences within each category, such as different play styles within the same roles (i.e., not all bigs, guards, or wings play the same way). Another limitation is that our analysis compares eras at a broad level instead of following the same players over time. For future research, it would be useful to break players into more specific positions, include more context such as minutes or team style, and look at player-level changes over time to better understand how the relationship between height and three-point shooting continues to evolve.
