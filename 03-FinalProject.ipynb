## Abstract

In order to understand how the relationship between NBA player height and three-point shooting behavior changed over time, we conducted an exploratory and statistical analysis covering three distinct eras of professional basketball. Our research was driven by the dramatic strategic transformation the NBA has undergone over the past 30 years; from a big man dominated era in the 1990s to the current analytics-driven game where three-point shooting is required of players at every position. In light of this change, we aimed to provide observational findings on whether height, which has long been thought to be a crucial factor in determining a player’s role and shot selection, still predicts how often a player attempts three-pointers and whether that relationship has changed over time. 

Using two endpoints, LeagueDashPlayerStats for season-level box score data and PlayerIndex for player height and positions metadata, we collected player statistics and biographical information from the official NBA Stats API using the nba_api Python module. After filtering for significant playing time, we acquired almost 6,000 player-season observations from 8 representative seasons across three eras (1990-2000, 2001-2013, and 2014-present). We selected FG3A_RATE, which is defined as three-point attempts as a percentage of all shot attempts, as our main variable of interest since it can represent shot selection independent of playing time and era-to-era pace differences. We grouped players into three positional categories (Guards, Wings, and Bigs) and applied linear regression within each era to determine the direction, strength, and statistical significance. 

Our analysis’s findings showed that player height and three-point attempt rate have a negative correlation that is highly statistically significant in each of the three eras. However, the league as a whole has shifted dramatically upward in three-point shooting over time. Most shockingly, from 1996 to 2024, FG3A_RATE increased by 1906% for Bigs (centers and power forwards) who historically attempted very few three-pointers, compared to around 200% for Guards and Wings. The absolute difference between shorter and taller players remained the same despite this, and the regression slope was consistently negative during all three eras since Guards and Wings also significantly raised their rates. 

Our study concludes by clarifying the evolving yet persistent relationship between professional basketball players’ height and three-point shooting. Although height is still a statistically significant predictor of three-point attempt rate, role, position, and era each play an important part on a player’s shot selection. Most notably, the modern NBA seems to be splitting, creating a new “stretch big” archetype who shoots from distance alongside traditional interior players who do not. This study contributes to the broader people of sports analytics and player role evaluation, showing insights for coaches, analysts, and fans interested in how the game continues to evolve. 

Taken together, these results suggest that the expansion of three-point shooting has not erased traditional size-based tendencies, but has instead layered new responsibilities onto existing player archetypes. Taller players remain less likely to rely on perimeter shooting than shorter players, yet the steady increase in long-range attempts across all position groups reflects a league-wide shift in offensive priorities. This pattern is especially relevant for understanding roster construction and player development, since it points to a modern game in which versatility and floor spacing are increasingly valued regardless of position. By showing both the persistence and the evolution of these trends, our project provides a clearer picture of how NBA shot selection has adapted over time.

```python
# Run this everytime you load this notebook so the necessary functions can be performed.

import os
import time
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
wrangled_df = pd.read_csv('data/02-processed/nba_height_3pa_player_season.csv')
wrangled_df.head()
```

##### Change #1: Replacement for Median FG3A Rate by Era and Position Group from previous checkpoint, was line plot before, now bar plot.

```python
by_era = wrangled_df.groupby(['ERA', 'POSITION_GROUP'])['FG3A_RATE'].median().reset_index()
by_era['ERA'] = pd.Categorical(by_era['ERA'], categories=['1990-2000', '2001-2013', '2014-present'])
by_era = by_era.sort_values('ERA')
sns.barplot(data=by_era, x='ERA', y='FG3A_RATE', hue='POSITION_GROUP')
plt.title('Median FG3A Rate by Era and Position Group')
plt.xlabel('Era')
plt.ylabel('Median FG3A Rate (FG3A / FGA)')
plt.legend(title='Position Group')
plt.show()
```

##### Change #2: Add p-values to the regression output, since previous checkpoitn never had that on the plots.

```python
# Display slopes of the regression lines for each era in the height vs. rate of 3-pointers attempted
for era in ['1990-2000', '2001-2013', '2014-present']:
    data = wrangled_df[wrangled_df['ERA'] == era]
    slope, intercept, r, p, se = stats.linregress(data['HEIGHT_IN'], data['FG3A_RATE'])
    print(f"{era}: slope={slope:.4f}, r= {r:.3f}, p= {p:.4f}, s= {se:.4f}")
```

The new previous display on checkpoint 2 only displayed the slope for each eras, now it displays the p-value and standard error as well. The extremely low p-values signify that the negative relationship between height and 3-pointer rates are highly significant in all 3 eras.

## Discussion and Conclusion

Overall, our results support the main claim stated in the abstract: player height remains negatively associated with three-point attempt rate in every era we examined, even as the NBA has become much more three-point oriented over time. In other words, taller players are still generally less likely to rely on three-point shooting than shorter players, but the modern game has raised three-point volume across the board. This is what makes our findings important. The league-wide increase in perimeter shooting did not eliminate the traditional height-based pattern; instead, it shifted the baseline upward for nearly everyone while preserving a negative relationship between height and FG3A_RATE.

The clearest example of this shift is the change among bigs. As noted in the abstract, FG3A_RATE increased by 1906% for bigs from 1996 to 2024, far more than the increase seen for guards and wings. This suggests that the modern NBA has expanded the offensive responsibilities of taller players without fully erasing the role differences that existed in earlier eras. Guards still tend to attempt threes at the highest rates, but wings and especially bigs have moved much further toward the perimeter, reflecting the growing importance of spacing, versatility, and outside shooting in roster construction and player development.

These results are consistent with prior basketball research showing that modern offenses emphasize three-point shooting more heavily than earlier eras and that positional boundaries have become less rigid over time. At the same time, our findings extend that idea by showing that the relationship can also be seen continuously through height, not only through listed positions. The more gradual slope in the 2014-present era suggests that the gap in three-point behavior between shorter and taller players may be narrowing, even though it has not disappeared. This supports the idea that the modern NBA is not replacing old player archetypes entirely, but rather layering new shooting expectations onto them.

There are also important limitations to this study. Because we used observational season-level data, we cannot conclude that height directly causes differences in three-point attempt behavior. Shot selection is also influenced by coaching strategy, team system, player skill, and offensive role. In addition, our position groups are broad and may hide meaningful differences within each category, while our era-based comparison does not track how the same individual players changed over time. Future work could address these issues by using more detailed role classifications, incorporating team-level context, or following players longitudinally across multiple seasons. Even with these limitations, our analysis provides strong evidence that height remains an important predictor of shot selection, while also showing how the modern NBA has reshaped the way players of every size participate in perimeter offense.
